## Joint Probabilites (and | or) - Multiple Columns

## what is the probability of playing cricket and temparature is hot?

```python
sum((df['Play']=='Yes') & (df['Temperature']=='Hot'))
```

```python
sum((df['Play']=='Yes') & (df['Temperature']=='Hot'))/len(df)
```

```python
pd.crosstab(index=df['Play'],columns=df['Temperature'],normalize=True).loc['Yes']['Hot']
```

```python
pd.crosstab(index=df['Temperature'],columns=df['Play'],normalize=True).loc['Hot']['Yes']
```

## Conditional Probability -- one event is happend, what is the chance of other event

## what is the probability of playing cricket if it is raining

```python
pd.crosstab(index=df['Play'],columns=df['Outlook'],normalize=True)
```

```python
pd.crosstab(index=df['Play'],columns=df['Outlook'],normalize='columns')
```

```python
pd.crosstab(index=df['Play'],columns=df['Outlook'],normalize='columns').loc['Yes']['Rain']
```

## what is the probablity of not playing cricket when humidity is high?

```python
pd.crosstab(index=df['Play'],columns=df['Humidity'],normalize='columns')
```

```python
pd.crosstab(index=df['Play'],columns=df['Humidity'],normalize='columns').loc['No']['High']
```

```python
pd.crosstab(index=df['Humidity'],columns=df['Play'],normalize='index').loc['High']['No']
```

```python
pd.crosstab(index=df['Humidity'],columns=df['Play'],normalize='index').loc['High']
```
